# 1. Top Violation Types

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=itv, y='Violation_Type', order=itv['Violation_Type'].value_counts().index, palette='Set2')
plt.title('Top Violation Types')
plt.xlabel('Count')
plt.ylabel('Violation Type')
plt.show()
```

#  2. Total Fine Collected by Violation Type

```python
fine_by_violation = itv.groupby('Violation_Type')['Fine_Amount'].sum().sort_values()
plt.figure(figsize=(12, 6))
fine_by_violation.plot(kind='barh', color='coral')
plt.title('Total Fine Amount by Violation Type')
plt.xlabel('Total Fine ($)')
plt.show()
```

# 3.Violation by Driver Gender

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=itv, x='Driver_Gender', palette='pastel')
plt.title('Violations by Driver Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

# 4.Violation Count by Vehicle Type

```python
plt.figure(figsize=(10, 5))
sns.countplot(data=itv, y='Vehicle_Type', order=itv['Vehicle_Type'].value_counts().index, palette='Set3')
plt.title('Violation Count by Vehicle Type')
plt.xlabel('Count')
plt.ylabel('Vehicle Type')
plt.show()
```

# 5.Violations by Time of Day

```python
itv['Hour'] = pd.to_datetime(itv['Time'], errors='coerce').dt.hour

plt.figure(figsize=(12, 6))
sns.histplot(data=itv, x='Hour', bins=24, kde=True, color='slateblue')
plt.title('Violations by Hour of the Day')
plt.xlabel('Hour')
plt.ylabel('Number of Violations')
plt.xticks(range(0, 24))
plt.show()
```

# 6.Effect of Road & Weather Conditions

```python
plt.figure(figsize=(14, 6))
sns.countplot(data=itv, x='Road_Condition', hue='Weather_Condition', palette='coolwarm')
plt.title('Violations by Road and Weather Conditions')
plt.xlabel('Road Condition')
plt.ylabel('Count')
plt.legend(title='Weather')
plt.show()
```

# 7.Fine Amount by Driver Age Group

```python
itv['Age_Group'] = pd.cut(itv['Driver_Age'], bins=[16, 25, 35, 50, 65, 100], labels=['16-25', '26-35', '36-50', '51-65', '66+'])

plt.figure(figsize=(10, 6))
sns.boxplot(data=itv, x='Age_Group', y='Fine_Amount', palette='spring')
plt.title('Fine Amount Distribution by Driver Age Group')
plt.xlabel('Driver Age Group')
plt.ylabel('Fine Amount')
plt.show()
```

# 8.Recorded Speed vs. Speed Limit

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=itv, x='Speed_Limit', y='Recorded_Speed', hue='Traffic_Light_Status', alpha=0.6)
plt.plot([itv['Speed_Limit'].min(), itv['Speed_Limit'].max()], 
         [itv['Speed_Limit'].min(), itv['Speed_Limit'].max()], 'r--')  # y=x line
plt.title('Recorded Speed vs. Speed Limit')
plt.xlabel('Speed Limit')
plt.ylabel('Recorded Speed')
plt.legend(title='Traffic Light')
plt.show()
```

 # 9. Helmet / Seatbelt Compliance

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

sns.countplot(data=itv, x='Helmet_Worn', ax=axs[0], palette='Set1')
axs[0].set_title('Helmet Worn Distribution')

sns.countplot(data=itv, x='Seatbelt_Worn', ax=axs[1], palette='Set2')
axs[1].set_title('Seatbelt Worn Distribution')

plt.tight_layout()
plt.show()
```

# 10. Top Locations with Most Violations

```python
top_locations = itv['Location'].value_counts().nlargest(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_locations.values, y=top_locations.index, palette='mako')
plt.title('Top 10 Violation Locations')
plt.xlabel('Number of Violations')
plt.ylabel('Location')
plt.show()
```

 # 11.Heatmap of Correlations

```python
plt.figure(figsize=(12, 8))
sns.heatmap(itv.select_dtypes(include='number').corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap (Numeric Columns)')
plt.show()
```
